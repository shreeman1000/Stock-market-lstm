```python
# importing the required libraries
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
from sklearn.model_selection import train_test_split
from urllib.request import urlopen
from goose3 import Goose
import re
from bs4 import BeautifulSoup as soup
g = Goose()
```

```python
# the 1st site is the times of india site
url = 'https://timesofindia.indiatimes.com/world'
conn1 = urlopen(url)
page_html = conn1.read()
conn1.close()
# soup is used to parse html data
html = soup(page_html,'html.parser')
```

```python
news_card = html.findAll('div',{'class':'news_card'})
```

```python
# the links to news articles are stores in span class = w_tle which is a part of the dic class news_card
toi_news = {}
t = 0
for i in news_card:
    for k,j in enumerate(i.findAll('span',{'class':'w_tle'})):
        try:
            toi_news[t*4+k] = 'https://timesofindia.indiatimes.com'+j.a['href']
        except:
            pass
    t += 1
```

```python
# using goose to extract and clean data and using .lower() to convert data to lower case
for i,temp_url in toi_news.items():
    a = g.extract(temp_url)
    toi_news[i] = a.cleaned_text.lower()
```

```python
toi_news
```

```python
# the second site chosen is Inshorts.com
is_url = 'https://inshorts.com/en/read'
is_conn = urlopen(is_url)
is_html = is_conn.read()
is_c = soup(is_html,'html.parser')
```

```python
# on this site the article content are straightforward stored in excerpts in div itemprop articleBody
y = is_c.findAll('div',{'itemprop': 'articleBody'})
```

```python
in_s = {}
for i,j in enumerate(y):
    in_s[i] = j.text.lower()
```

```python
in_s
```

```python
# the third site chosen is newyorktimes
url = 'https://www.nytimes.com/section/world'
conn = urlopen(url)
page_html = conn.read()
html = soup(page_html,'html.parser')
```

```python
# all articles are stored in h2
a = html.findAll('h2')
```

```python
# using a try except loop to extract link for articles and check whether the found h2 is actually a link or not
```

```python
counter=0
ny={}
for i in a:
    try:
        ny[counter] = 'https://www.nytimes.com'+i.a['href']
        counter += 1
    except:
        pass
```

```python
for i,temp_url in ny.items():
    a = g.extract(temp_url)
    ny[i] = a.cleaned_text.lower()
```

```python
ny
```

```python
# removing \n and \' which are prevelant in data
for i in range(len(ny)):
    ny[i] = re.sub('[\n]', '', ny[i])
    ny[i] = re.sub('[\']', '', ny[i])
```

```python
for i in range(len(toi_news)):
    toi_news[i] = re.sub('[\n]', '', toi_news[i])
    toi_news[i] = re.sub('[\']', '', toi_news[i])
```

```python
for i in range(len(in_s)):
    in_s[i] = re.sub('[\n]', '', in_s[i])
    in_s[i] = re.sub('[\']', '', in_s[i])
```

```python
# nytimes offers some articles in spanish
# removing the articles in spanish
ny.pop(15)
ny.pop(16)
ny.pop(17)
ny.pop(18)
ny.pop(19)
ny.pop(20)
ny
```

```python
df2.rename(columns = {'@switchfoot http://twitpic.com/2y1zl - Awww, that\'s a bummer.  You shoulda got David Carr of Third Day to do it. ;D':'TEXT'}, inplace = True)
```

```python
df2.rename(columns = {'0' : 'Result'}, inplace = True)
```

```python
# cleaning the dataset by removing links @usertags and other useless thind like emojis and replacing them
# wiht appropiate sentiment or empty value
urlPattern        = r"((http://)[^ ]*|(https://)[^ ]*|( www\.)[^ ]*)"
userPattern       = '@[^\s]+'
sequencePattern   = r"(.)\1\1+"
seqReplacePattern = r"\1\1"
```

```python
emojis = {':)': 'smile', ':-)': 'smile', ';d': 'wink', ':-E': 'vampire', ':(': 'sad', 
          ':-(': 'sad', ':-<': 'sad', ':P': 'raspberry', ':O': 'surprised',
          ':-@': 'shocked', ':@': 'shocked',':-$': 'confused', ':\\': 'annoyed', 
          ':#': 'mute', ':X': 'mute', ':^)': 'smile', ':-&': 'confused', '$_$': 'greedy',
          '@@': 'eyeroll', ':-!': 'confused', ':-D': 'smile', ':-0': 'yell', 'O.o': 'confused',
          '<(-_-)>': 'robot', 'd[-_-]b': 'dj', ":'-)": 'sadsmile', ';)': 'wink', 
          ';-)': 'wink', 'O:-)': 'angel','O*-)': 'angel','(:-D': 'gossip', '=^.^=': 'cat'}
```

```python
# applying cleaning and storing the data in a list
ptext = []
for i in df2['TEXT'].values:
    i = i.lower()
    i = re.sub(urlPattern,' URL',i)
    for emoji in emojis.keys():
        i = i.replace(emoji, "EMOJI" + emojis[emoji])        
    i = re.sub(userPattern,'USER', i)
    i = re.sub(sequencePattern, seqReplacePattern, i)
    ptext.append(i)
```

```python
len(ptext)
```

```python
ptext
```

```python
df_cleantext = pd.DataFrame(ptext,columns = ['Clean'])
```

```python
df_cleantext
```

```python
df_clean = df2.join(df_cleantext)
```

```python
df_clean.drop('TEXT',axis = 1,inplace =True)
```

```python
df_clean
```

```python
X = df_clean['Clean'].values
# 0 - negative sentiment , 4 - positive sentiment
y = df_clean['Result'].values
```

```python
# vectorizer is a encoding algorithm used to make columns based on words in sentence
# like one hot encoding for columns with sentences
from sklearn.feature_extraction.text import TfidfVectorizer
vectoriser = TfidfVectorizer(ngram_range=(1,2), max_features=500000)
```

```python
vectoriser.fit(X)
```

```python
X = vectoriser.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.1, random_state = 0)
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
